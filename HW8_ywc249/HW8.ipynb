# 2. Data Cleaning
## 2.1 Turnstile Data

```python
# read data
TURN = pd.read_csv(os.getenv('PUIDATA') + '/mta1028.txt')
TURN.head()
```

```python
# subset data for stations on D line only
# group by aggregations of entries and exits per station
dline = ['D' in l for l in TURN.LINENAME]
TURN2 = TURN[dline].groupby(by=TURN.STATION).sum().reset_index()
TURN2.head()
print(TURN2.shape)
```

## 2.3 Merge Data & Data Engineering

```python
# merge data by station
# turnstile data now includes station attributes
df = pd.merge(TURN2, STN2, left_on='STATION', right_on='NAME', indicator=True)
df
print(df.shape)

### NOTE: 3 stations are dropped out, see '4. Next Update' for explanation ###
```

```python
# fix corrupted column name
df.rename(columns={df.columns[2]:'EXITS'}, inplace=True)

# calculate net passenger flow for each station
# transformed into percentage based on entries
# 0% means entries equal to exits
# negative percentage means more people exit from the station than enter the station
df['DIFF'] = (df.ENTRIES - df.EXITS) / df.ENTRIES * 100
```

```python
# extract Latitude data from 'the_geom'
df['LAT'] = [c.split(' ')[2][:-1] for c in df.the_geom]

# sort stations by latitude
# for a horizontal bar plot, sort ascendingly
# for a vertical bar plot, sort descendingly
df.sort_values(by='LAT', ascending=True, inplace=True)

# reset index
df.reset_index(inplace=True)
df.drop('index', axis=1, inplace=True)
df
```

# 3. Visualization

```python
# set colors
# bar colors distincted by positive and negative percentages
barcolor=df.DIFF.apply(lambda x:'steelblue' if x>=0 else 'indianred')
# font colors made white when negative bars highlight too much of the station names
fontcolor=df.DIFF.apply(lambda x:'black' if x>=-100 else 'white')

# plot the horizontal bar plot
plt.figure(figsize=(10,12))
ax = df.DIFF.plot(kind='barh', width=0.7, color=barcolor)

# set x-ticks: percentage
# take 25-percent intervals between -100 to 100
# take longer intervals beyond -100 to accommodate 'GRAND ST' station
# place x-ticks labels on top of the figure
xticks = list(range(-400, -100, 200)) + list(range(-100, 101, 25))
ax.set_xticks(xticks)
ax.tick_params(axis='x', pad=2, labelsize=14, labeltop=True, labelbottom=False)

# set y-ticks: station name
# use annotation instead of y-ticks
ax.set_yticks([])
for i,s in enumerate(df.STATION):
    ax.annotate('%s'%(s), xy=(0,i), xytext=(-2,-3.5), textcoords='offset points',
                rotation=0, fontsize=12, color=fontcolor[i], horizontalalignment='right')

# remove horizontal gridlines
ax.grid(False, axis=0)

# add a compass
# annotate north and add an arrow pointing to it
ax.annotate('North', xy=(-350,35), xytext=(-30,10), textcoords='offset points',
            rotation=0, fontsize=30)
ax.arrow(-345, 32.5, 0, 1.5, width=2, head_width=15, head_length=1, fc='k', ec='k')

# add a D line mark
circ = mpatches.Circle((210, 750), 50, color='#FF6319', alpha=0.8, transform=None)
ax.add_artist(circ)
ax.annotate('D', xy=(-260,33), xytext=(-36,6), textcoords='offset points',
            rotation=0, fontsize=70, color='white')

# add labels and title
# place x-ticks labels on top of the figure
ax.set_xlabel('Percentage of Net Passenger Flow (%)', fontsize=18, labelpad=8)
ax.set_ylabel('D Line Stations', fontsize=18, labelpad=5)
ax.set_title('', fontsize=24)
ax.xaxis.set_label_position('top')

# add metadata
ax.annotate('Yuwen Chang (ywc249)', xy=(-390,0), rotation=0, fontsize=13)
ax.annotate('Updated: 2017/11/11', xy=(-390,0.7), rotation=0, fontsize=13)

# finalize the figure and export it to a png file and onto the screen
plt.tight_layout()
plt.savefig('HW8.png')
plt.show()
```
