```python
import os
import numpy as np
import pandas as pd
```

```python
# Data 1: The Guardian firearms murders and ownership data
link1 = 'https://docs.google.com/spreadsheets/d/1chqUZHuY6cXYrRYkuE0uwXisGaYvr7durZHJhpLGycs/edit#gid=0'
data1 = pd.read_csv(os.getenv("PUIDATA") + "/" + "firearm.csv")

# Data 2: Number of mass shootings
link2 = 'https://datawrapper.dwcdn.net/pvLFI/4/'
data2 = pd.read_csv(os.getenv("PUIDATA") + "/" + "massshoot.csv")

# Data 3: The world population by country
link3 = 'http://api.worldbank.org/v2/en/indicator/SP.POP.TOTL?downloadformat=csv'
data3 = pd.read_csv(os.getenv("PUIDATA") + "/" + "worldpop.csv", header=2)
```

# Merge 1 & 2

```python
data12 = pd.merge(data1, data2, left_on="Country/Territory", right_on="Country", how='outer', indicator=True)
data12
```

```python
# data1 (left) should have a more complete list of countries
# check 'right_only' to see potential naming issues
data12[data12._merge == 'right_only']
```

```python
# search country using 2-digit iso code
iso = ['GB', 'UK', 'NL', 'BE']
for i in iso:
    print(data1['Country/Territory'][data1['ISO code'] == i])
```

```python
# United Kingdom is broken into different jurisdictions
# It is possible to join an recalculate part of the data to GB as a whole but I am dropping it

# rename country for consistency
data2.iloc[data2[data2['Country'] == 'Beligium'].index,0] = 'Belgium'
data2.iloc[data2[data2['Country'] == 'Holland'].index,0] = 'Netherlands'

# merge again
data12 = pd.merge(data1, data2, left_on="Country/Territory", right_on="Country", how='outer', indicator=True)
data12[data12._merge == 'right_only']
```

# Merge 12 & 3

```python
data12.drop('_merge', axis=1, inplace=True)

data123 = pd.merge(data12, data3, left_on="Country/Territory", right_on="Country Name", how='outer', indicator=True)
data123
```

```python
# Too many right_only merges
# export country namelist for external review
data1ex = data1['Country/Territory']
data3ex = data123['Country Name'][data123._merge == 'right_only']

dftemp = pd.DataFrame(data=[list(data1ex),list(data3ex)]).T
dftemp.to_csv('countrylist.csv')
```

```python
# examine gibberish item
data1[data1['ISO code'] == 'CI']
```

```python
# fix gibberish
data1.iloc[data1[data1['ISO code'] == 'CI'].index,0] = "Cote d'Ivoire"
data1[data1['ISO code'] == 'CI']
```

```python
# read in the map file for renaming
dftemp2 = pd.read_csv(os.getenv("PUIDATA") + "/" + "countrylist.csv", index_col=0)
dftemp2
```

```python
# rename country name for consistency
d_country = {key: value for (key, value) in zip(dftemp2['country_old'], dftemp2['country_new'])}
data3.replace({"Country Name": d_country}, inplace=True)
data3
```

# Redo merge-all

```python
data12 = pd.merge(data1, data2, left_on="Country/Territory", right_on="Country", how='outer')
data123 = pd.merge(data12, data3, left_on="Country/Territory", right_on="Country Name", how='outer', indicator=True)
data123

#data123['Country Name'][data123._merge == 'right_only']
```
